```python
title=soup.find_all('div' ,class_="heading_4_5 profile")
title
```

```python
job_titles= []
for i in title:
    job_titles.append(i.text)
job_titles
```

```python
job_titles= []
for i in title:
    job_titles.append(i.text.replace('\n', ''))
job_titles
```

```python
companies=soup.find_all('a' ,class_="link_display_like_text")
companies
```

```python
company_names= []
for i in companies:
    company_names.append(i.text.strip())
company_names
```

```python
ctc=soup.find_all('i' ,class_="ic-16-currency-inr")
ctc
```

```python
ctc_list= []
for i in ctc:
    ctc_list.append(i.text)
ctc_list
```

```python
print(len(job_titles), len(company_names))
```

```python
jobs=pd.DataFrame({})
jobs['titles']=job_titles
jobs['company']=company_names
```
